```python
import csv
import requests
from bs4 import BeautifulSoup
import json
```

```python
csv_file_path = 'University List - Sheet1.csv'
output_json_path = 'output.json'
universities_info = {}

with open(csv_file_path, mode = 'r', encoding = 'utf-8') as csv_file:
    csv_reader = csv.reader(csv_file)
    next(csv_reader)

    for row in csv_reader:
        if not all(row[:3]):
            continue

        univ_name, admission_url, faq_url = row[:3]
        universities_info[univ_name] = {"FAQs": [], "AdmissionText": ""}

        try:
          faq_response = requests.get(faq_url, headers={'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'}, verify = False)

          if faq_response.status_code == 200:
                soup = BeautifulSoup(faq_response.text, 'html.parser')
                texts = soup.find_all(text = True)
                faqs = [text.strip() for text in texts if text.strip().endswith('?')][:10]
                universities_info[univ_name]["FAQs"] = faqs

        except Exception as e:
            print(f"Error fetching FAQ for {univ_name}: {e}")

        try:
          admission_response = requests.get(admission_url, headers = {'User-Agent': 'Mozilla/5.0'}, verify = False)

          if admission_response.status_code == 200:
              soup = BeautifulSoup(admission_response.text, 'html.parser')
              admission_texts = soup.find_all('p')
              admission_paragraphs = [text.get_text(strip=True) for text in admission_texts if text.get_text(strip = True)]
              universities_info[univ_name]["AdmissionText"] = admission_paragraphs

        except Exception as e:
            print(f"Error fetching admission info for {univ_name}: {e}")

with open(output_json_path, 'w', encoding = 'utf-8') as json_file:
    json.dump(universities_info, json_file, ensure_ascii = False, indent = 4)
```

```python
json_file_path = 'output.json'
csv_file_path = 'data.csv'

with open(json_file_path, 'r', encoding = 'utf-8') as file:
    universities_info = json.load(file)

with open(csv_file_path, mode = 'w', encoding = 'utf-8', newline = '') as file:
    csv_writer = csv.writer(file)
    csv_writer.writerow(['University Name', 'Admission Text', 'FAQs'])

    for univ_name, info in universities_info.items():
        faqs_combined = " | ".join(info["FAQs"])
        admission_text_combined = " ".join(info["AdmissionText"]) if isinstance(info["AdmissionText"], list) else info["AdmissionText"]
        csv_writer.writerow([univ_name, admission_text_combined, faqs_combined])
```
